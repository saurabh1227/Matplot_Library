**1. Load the necessary package for plotting using pyplot from matplotlib. Example - Days(x-axis) represents 8 days and Speed represents a car’s speed. Plot a Basic line plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.**
         Days=[1,2,3,4,5,6,7,8]

         Speed=[60,62,61,58,56,57,46,63]​​​​​

```python
import matplotlib.pyplot as plt

days = [1, 2, 3, 4, 5, 6, 7, 8]
car_speed = [60, 62, 61, 58, 56, 57, 46, 63]
plt.plot(days, car_speed)
plt.xlabel("Days")
plt.ylabel("Car speed in km/h")
plt.title("care speed day by day")
plt.show()
```

**3. Plot Axes Labels, Chart title, Legend, Grid in Car minimum, Maximum and average speed in 8 days.**

```python
days = [1, 2, 3, 4, 5, 6, 7, 8] 
max_speed = [80, 91, 92, 88, 77, 79, 76, 75] 
min_speed = [42, 43, 40, 42, 33, 36, 34, 35]
avg_speed = [46, 58, 57, 56, 40, 42, 41, 36]

plt.title("Maximum, minimum and average speed of cars")
plt.xlabel("Days")
plt.ylabel("Speed in km/h")
plt.plot(days, max_speed, color = "red", label = "Maximum speed")
plt.plot(days, min_speed, color = "green", label = "Minimum speed")
plt.plot(days, avg_speed, color = "blue", label = "Average speed")
plt.legend()
plt.grid()
plt.show()
```

**4. Plotting a basic sine graph by adding more features. Adding Multiple plots by Superimposition like cosine wave.**

```python
import numpy as np

#sine wave
time = np.arange(0, 10, 0.2)
amplitude = np.sin(time)
amplitude1 = np.cos(time)
plt.plot(time, amplitude, color = "cyan", label = "sine wave")
plt.plot(time, amplitude1, color = "purple", label = "cosine wave")
plt.grid()
plt.legend()
plt.show()
```

**5. Plot Simple bar chart showing popularity of Programming Languages.**

Languages =['Python', 'SQL', 'Java', 'C++', 'JavaScript']

Popularity = [56, 39, 34, 34, 29]

Security = [44 ,36 ,55, 50, 42]

**Plot Multiple Bars showing Popularity and Security of major Programming Languages. Also Create Horizontal bar chart using barh function.**

```python
languages = ["Python", "SQL", "Java", "C++", "JavaScript"]
popularity = [56, 39, 34, 34, 29]
security = [44, 36, 55, 50, 42]

plt.bar(languages, popularity, color = "grey", width = 0.7, label = "Popularity")
plt.bar(languages, security, color = "black", width = 0.5, label = "Security")
plt.legend()
plt.title("Popularity of programming languages")
plt.xlabel("Languages")
plt.ylabel("Popularity and Security")
plt.show()
```

```python
plt.barh(languages, popularity, label = "Popularity", height = 0.8)
plt.barh(languages, security, label = "Security", height = 0.5)
plt.legend()
plt.show()
```

**7. Titanic Data Set Download Data**

 Load the data file

 (i) Create a pie chart presenting the male/female proportion 

 (ii) Create a scatterplot with the Fare paid and the Age, differ the plot color by gender

```python
import pandas as pd
titanic = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/train.csv")
titanic.head()
```

```python
# pie chart for male-female proportion
sex = titanic.groupby(["Sex"])["Survived"].count()
print(sex)
plt.pie(sex, labels = sex.index, startangle = 90, counterclock = False)
plt.legend()
plt.title("Survived sex ratio in titanic")
plt.show()
```

```python
# Create a scatterplot with the Fare paid and the Age, differ the plot color by gender.

male = titanic[titanic["Sex"] == "male"]
print(male.Sex.value_counts())
female = titanic[titanic["Sex"] == "female"]
print(female.Sex.value_counts())
plt.scatter(male.Age, male.Fare, label = "Male")
plt.scatter(female.Age, female.Fare, label = "Female")
plt.title("Fare paid and age")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend()
plt.show()
```
